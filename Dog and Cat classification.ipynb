```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import random
import keras
import tensorflow
import cv2
```

```python
DIRECTORIES=r'C:\Users\ASUS\Desktop\Dog and Cat Dataset\train'
CATAGORIES=['Cat','Dog']
```

```python
daata=[]
for catagories in CATAGORIES:
    folder=os.path.join(DIRECTORIES,catagories)
    label=CATAGORIES.index(catagories)
    print(label)
    for img in os.listdir(folder):
        img_path=os.path.join(folder,img)
        img_array=cv2.imread(img_path)
        img_g=cv2.resize(img_array,(100,100))
        daata.append([img_g,label])
        
```

```python
daata
```

```python
len(daata)
```

```python
random.shuffle(daata)
```

```python
X=[]
y=[]
for features,labels in daata:
    X.append(features)
    y.append(labels)
```

```python
X
```

```python
y
```

```python
x=np.array(X,dtype=np.float32)
y=np.array(y,dtype=np.float32)
```

```python
x
```

```python
x=x/255
x
```

```python
x.shape
```

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
```

```python
Model = Sequential()
Model.add(Conv2D(64,(3,3),activation='relu'))
Model.add(MaxPooling2D(2,2))

Model.add(Conv2D(64,(3,3),activation='relu'))
Model.add(MaxPooling2D(2,2))

Model.add(Flatten())

Model.add(Dense(128,input_shape=(100, 100, 3),activation='relu'))

Model.add(Dense(2,activation='softmax'))
```

```python
Model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
Model.fit(x,y,epochs=10,validation_split=0.1)
```

```python
import pickle as pk
```

```python
with open('DogCatModel','wb') as file:
    pk.dump(Model,file)
```
